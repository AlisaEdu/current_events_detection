```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

df = pd.DataFrame(columns=['date, text, link'])
# print(i)
for i in ['Styczeń_2022', 'Luty_2022', 'Marzec_2022', 'Kwiecień_2022', 'Maj_2022', 'Czerwiec_2022', 'Lipiec_2022', 'Sierpień_2022', 'Wrzesień_2022', 'Październik_2022', 'Listopad_2022', 'Grudzień_2022']:
  # Make a request to the URL
  url = f'https://pl.wikipedia.org/wiki/{i}'
  response = requests.get(url)

  # Parse the HTML content using BeautifulSoup
  soup = BeautifulSoup(response.content, 'html.parser')

  # Find all the h2 elements
  h2_tags = soup.find_all('h2')

  # Initialize lists to store the data
  dates = []
  news_items = []
  links = []

  # Loop through each h2 element
  for h2 in h2_tags[1:-1]:
      # Get the date from the h2 text
      date = h2.text.strip()

      # Find the next ul element after the h2
      ul = h2.find_next_sibling('ul')

      # Loop through each li element in the ul
      for li in ul.find_all('li'):
          # Get the news item text
          news_item = li.text.strip()

          # Get all links in the li element
          # for a in li.find_all('a'):
          link = li.find_all('a')[0].get('href')
          if link and link.startswith('/wiki/'):
              dates.append(date)
              news_items.append(news_item)
              links.append(link)
              

  # Create a DataFrame with the data
  df_temp = pd.DataFrame({
      'Date': dates,
      'News Item': news_items,
      'Link': links
  })

  df = df.append(df_temp)
```
